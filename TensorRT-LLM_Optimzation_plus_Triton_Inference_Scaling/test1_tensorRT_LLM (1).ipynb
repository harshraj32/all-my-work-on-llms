```python
# For capturing output from the summarize commands
import re
import matplotlib.pyplot as plt
import subprocess

# Function to run command and capture output
def run_command(cmd):
    process = subprocess.Popen(cmd, stdout=subprocess.PIPE, stderr=subprocess.PIPE, shell=True, text=True)
    stdout, stderr = process.communicate()
    return stdout + stderr
```

```python
# Function to parse results
def parse_output(output):
    # Extract execution time
    time_match = re.search(r'real\s+(\d+)m([\d.]+)s', output)
    if time_match:
        minutes = int(time_match.group(1))
        seconds = float(time_match.group(2))
        exec_time = minutes * 60 + seconds
    else:
        exec_time = None

    # Extract ROUGE scores
    rouge_scores = re.findall(r'rouge[12Lsum]+ : ([\d.]+)', output)
    rouge_scores = [float(score) for score in rouge_scores] if rouge_scores else []

    # Extract performance metrics
    latency_match = re.search(r'total latency: ([\d.]+) sec', output)
    latency = float(latency_match.group(1)) if latency_match else None

    tokens_match = re.search(r'total output tokens: (\d+)', output)
    total_tokens = int(tokens_match.group(1)) if tokens_match else None

    tokens_per_sec_match = re.search(r'tokens per second: ([\d.]+)', output)
    tokens_per_sec = float(tokens_per_sec_match.group(1)) if tokens_per_sec_match else None

    return exec_time, rouge_scores, latency, total_tokens, tokens_per_sec
```

```python
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for more appealing visuals
sns.set(style="whitegrid")

# Improved function to parse results
def parse_output(output):
    # Extract execution time - checking multiple patterns
    time_match = re.search(r'real\s+(\d+)m([\d.]+)s', output)
    if time_match:
        minutes = int(time_match.group(1))
        seconds = float(time_match.group(2))
        exec_time = minutes * 60 + seconds
    else:
        # Try HF specific total latency pattern
        hf_latency = re.search(r'Hugging Face \(total latency: ([\d.]+) sec\)', output)
        if hf_latency:
            exec_time = float(hf_latency.group(1))
        else:
            # Try alternative pattern for execution time
            alt_time_match = re.search(r'latency: ([\d.]+) sec', output)
            if alt_time_match:
                exec_time = float(alt_time_match.group(1))
            else:
                # Try the format from your specific output
                model_latency = re.search(r'model latency: ([\d.]+) seconds', output)
                exec_time = float(model_latency.group(1)) if model_latency else None

    # Extract ROUGE scores - updated pattern to match your output format
    rouge_scores = []

    # First try the HF specific format with "rouge1 : 19.483..." pattern
    rouge1 = re.search(r'rouge1\s*:\s*([\d.]+)', output)
    rouge2 = re.search(r'rouge2\s*:\s*([\d.]+)', output)
    rougeL = re.search(r'rougeL\s*:\s*([\d.]+)', output)
    rougeLsum = re.search(r'rougeLsum\s*:\s*([\d.]+)', output)

    if rouge1 and rouge2 and rougeL and rougeLsum:
        rouge_scores = [
            float(rouge1.group(1)),
            float(rouge2.group(1)),
            float(rougeL.group(1)),
            float(rougeLsum.group(1))
        ]

    # If still empty, check for the array pattern from your output
    if not rouge_scores:
        array_match = re.search(r'ROUGE scores: \[([\d., ]+)\]', output)
        if array_match:
            try:
                # Convert string representation of array to actual list of floats
                score_str = array_match.group(1)
                rouge_scores = [float(s.strip()) for s in score_str.split(',') if s.strip()]
            except:
                pass

    # If still empty, look for specific array pattern in your example
    if not rouge_scores:
        array_match = re.search(r'\[([\d.]+),\s*([\d.]+),\s*([\d.]+),\s*([\d.]+)', output)
        if array_match:
            try:
                rouge_scores = [float(array_match.group(i)) for i in range(1, 5)]
            except:
                pass

    # Extract performance metrics
    latency_match = re.search(r'total latency: ([\d.]+) sec', output)
    if not latency_match:
        latency_match = re.search(r'model latency: ([\d.]+) seconds', output)
    latency = float(latency_match.group(1)) if latency_match else None

    tokens_match = re.search(r'total output tokens: (\d+)', output)
    if not tokens_match:
        tokens_match = re.search(r'output tokens: (\d+)', output)
    total_tokens = int(tokens_match.group(1)) if tokens_match else None

    tokens_per_sec_match = re.search(r'tokens per second: ([\d.]+)', output)
    tokens_per_sec = float(tokens_per_sec_match.group(1)) if tokens_per_sec_match else None

    return exec_time, rouge_scores, latency, total_tokens, tokens_per_sec

# Use the actual HF output data
llama_hf_output = """
[03/03/2025-02:49:54] [TRT-LLM] [I] Hugging Face (total latency: 52.09791946411133 sec)
[03/03/2025-02:49:54] [TRT-LLM] [I] Hugging Face (total output tokens: 1235)
[03/03/2025-02:49:54] [TRT-LLM] [I] Hugging Face (tokens per second: 23.70536122561965)
[03/03/2025-02:49:54] [TRT-LLM] [I] HF beam 0 result
[03/03/2025-02:49:54] [TRT-LLM] [I]   rouge1 : 19.483494464993218
[03/03/2025-02:49:54] [TRT-LLM] [I]   rouge2 : 5.217070104817541
[03/03/2025-02:49:54] [TRT-LLM] [I]   rougeL : 14.667254819471504
[03/03/2025-02:49:54] [TRT-LLM] [I]   rougeLsum : 15.968486084892792
"""

# Keep INT8 output the same
llama_int8_output = """INT8 model execution time: None seconds
INT8 model ROUGE scores: [19.81987963993911, 6.444742615009022, 15.118948537402282, 16.02743861576778]
INT8 model latency: 11.633677244186401 seconds
INT8 model tokens per second: 102.54711171363876"""

# Parse the outputs
llama_hf_exec_time, llama_hf_rouge, llama_hf_latency, llama_hf_tokens, llama_hf_tps = parse_output(llama_hf_output)
llama_int8_exec_time, llama_int8_rouge, llama_int8_latency, llama_int8_tokens, llama_int8_tps = parse_output(llama_int8_output)

# If execution time is None, use latency instead
if llama_hf_exec_time is None and llama_hf_latency is not None:
    llama_hf_exec_time = llama_hf_latency
if llama_int8_exec_time is None and llama_int8_latency is not None:
    llama_int8_exec_time = llama_int8_latency

# Limit ROUGE scores to first 4 if there are duplicates (as seen in your output)
if len(llama_int8_rouge) > 4:
    llama_int8_rouge = llama_int8_rouge[:4]

# Create pandas DataFrame for visualization
model_names = ['HF', 'INT8']
data = {
    'Model': model_names + model_names + model_names + model_names + model_names,
    'Metric': ['Latency'] * 2 + ['Tokens/sec'] * 2 + ['ROUGE-1'] * 2 + ['ROUGE-2'] * 2 + ['ROUGE-Lsum'] * 2,
    'Value': [
        llama_hf_exec_time, llama_int8_exec_time,
        llama_hf_tps, llama_int8_tps,
        llama_hf_rouge[0], llama_int8_rouge[0],
        llama_hf_rouge[1], llama_int8_rouge[1],
        llama_hf_rouge[3], llama_int8_rouge[3]
    ]
}

df = pd.DataFrame(data)

# For better visualization, create a separate dataframe for latency
latency_df = df[df['Metric'] == 'Latency'].copy()
tps_df = df[df['Metric'] == 'Tokens/sec'].copy()
rouge_df = df[df['Metric'].str.startswith('ROUGE')].copy()

# Create a more visually appealing plot with Seaborn
plt.figure(figsize=(18, 12))

# Define a custom color palette
palette = {"HF": "#3498db", "INT8": "#e74c3c"}

# Plot 1: Latency
plt.subplot(2, 2, 1)
ax1 = sns.barplot(x='Model', y='Value', data=latency_df, palette=palette)
plt.title('Latency Comparison (lower is better)', fontsize=14, fontweight='bold')
plt.ylabel('Seconds', fontsize=12)
plt.xlabel('')

# Add value labels on top of bars
for i, p in enumerate(ax1.patches):
    ax1.annotate(f'{p.get_height():.2f}s',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom', fontsize=11, fontweight='bold')

# Plot 2: Tokens per second
plt.subplot(2, 2, 2)
ax2 = sns.barplot(x='Model', y='Value', data=tps_df, palette=palette)
plt.title('Tokens per Second (higher is better)', fontsize=14, fontweight='bold')
plt.ylabel('Tokens/sec', fontsize=12)
plt.xlabel('')

# Add value labels on top of bars
for i, p in enumerate(ax2.patches):
    ax2.annotate(f'{p.get_height():.2f}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom', fontsize=11, fontweight='bold')

# Plot 3: ROUGE scores
plt.subplot(2, 1, 2)
ax3 = sns.barplot(x='Metric', y='Value', hue='Model', data=rouge_df, palette=palette)
plt.title('ROUGE Score Comparison (higher is better)', fontsize=14, fontweight='bold')
plt.ylabel('Score', fontsize=12)
plt.xlabel('')
plt.legend(title='Model', loc='upper right')

# Add value labels on top of bars
for i, p in enumerate(ax3.patches):
    ax3.annotate(f'{p.get_height():.2f}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom', fontsize=9)

# Performance improvement annotations
speedup = llama_hf_exec_time / llama_int8_exec_time
tps_improvement = llama_int8_tps / llama_hf_tps

# Add a text box with performance summary
plt.figtext(0.5, 0.01,
           f"Performance Summary:\n"
           f"INT8 is {speedup:.2f}x faster than HF with {tps_improvement:.2f}x better token throughput\n"
           f"ROUGE scores are comparable between models, indicating minimal quality degradation with quantization",
           ha="center", fontsize=12, bbox={"facecolor":"lightgray", "alpha":0.5, "pad":5})

plt.tight_layout(rect=[0, 0.05, 1, 0.95])
plt.suptitle('LLaMA Model Performance Comparison: HF vs INT8', fontsize=16, fontweight='bold', y=0.99)
plt.subplots_adjust(top=0.9)

plt.show()
```
